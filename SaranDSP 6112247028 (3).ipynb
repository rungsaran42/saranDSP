```python
#Lab01

from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display

sin_sig = thinkdsp.SinSignal(freq=7028, amp=28, offset=0)
sin_sig.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#Lab02
signal = thinkdsp.SinSignal(28000)
duration = signal.period*280
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
snwave = signal.make_wave(duration=28, framerate=10000)
snwave.apodize()
snwave.make_audio()
```

```python
signal = thinkdsp.SquareSignal(2800)
duration = signal.period*280
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
sqwave = signal.make_wave(duration=28, framerate=10000)
sqwave.apodize()
sqwave.make_audio()
```

```python
spectrum= snwave.make_spectrum()
spectrum.plot()
```
